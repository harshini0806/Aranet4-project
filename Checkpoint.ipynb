```python
import os
import pandas as pd

def filter_csv_files(filenames):
    """Filter out non-CSV files from a list of filenames."""
    return [f for f in filenames if f.endswith('.csv')]

def read_csv_files_to_dataframes(csv_files, prefix='df_'):
    """Read CSV files into Pandas dataframes with a generic naming pattern."""
    dataframes = {}
    for i, filename in enumerate(csv_files):
        df_name = f'{prefix}{i}'
        dataframes[df_name] = pd.read_csv(filename)
    return dataframes

# List all files in the current directory
all_files = os.listdir('.')

# Filter out non-CSV files
csv_files = filter_csv_files(all_files)

# Read CSV files into dataframes
dataframes = read_csv_files_to_dataframes(csv_files)

# Print the names of the created dataframes
print("Created dataframes:", list(dataframes.keys()))

# Access a specific dataframe (e.g., the first one)
print(dataframes['df_0'].head())
```

```python

def clean_datetime_column(df, column_name):
    # Convert the column to datetime format
    df[column_name] = pd.to_datetime(df[column_name], format='%d/%m/%Y %I:%M:%S %p')
    
    # Create separate 'Date' and 'Time' columns in the desired formats
    df['Date'] = df[column_name].dt.strftime('%m/%d/%Y')
    df['Time'] = df[column_name].dt.strftime('%H:%M:%S')
    
    # Drop the original column
    df.drop(column_name, axis=1, inplace=True)
    return df

# Clean the date-time column
Aranet = clean_datetime_column(Aranet, 'Time(DD/MM/YYYY h:mm:ss A)')

Aranet.head(10)
```
